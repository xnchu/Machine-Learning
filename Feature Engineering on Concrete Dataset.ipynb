# Feature Engineering on Concrete Dataset

```python
import pandas as pd 
```

```python
df=pd.read_excel("Concrete_Data.xls")
```

```python
df.head()
```

```python
df.columns=['Cement','Blast Furnace','Fly Ash','Water','Superplasticizer','Coarse Aggregate','Fine Aggregate','Age','Compressive Strength']
```

```python
df.head()
```

```python
X=df.drop('Compressive Strength',axis=1)

y=df['Compressive Strength']
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score


# Train and score baseline model

baseline = RandomForestRegressor(criterion="mae", random_state=0)

baseline_score = cross_val_score(baseline, X, y, cv=5, scoring="neg_mean_absolute_error")
baseline_score = -1 * baseline_score.mean()

print(f"MAE Baseline Score: {baseline_score:.4}")
```

```python
X_1 = df.copy()
y_1 = X_1.pop("Compressive Strength")

# Create synthetic features
X_1["FCRatio"] = X_1["Fine Aggregate"] / X_1["Coarse Aggregate"]
X_1["AggCmtRatio"] = (X_1["Coarse Aggregate"] + X_1["Fine Aggregate"]) / X_1["Cement"]
X_1["WtrCmtRatio"] = X_1["Water"] / X_1["Cement"]

# Train and score model on dataset with additional ratio features
model = RandomForestRegressor(criterion="mae", random_state=0)

score = cross_val_score(model, X_1, y_1, cv=5, scoring="neg_mean_absolute_error")
score = -1 * score.mean()

print(f"MAE Score with Ratio Features: {score:.4}")
```
